```python
# Extended data for Egypt and Nigeria (2017-2023)
egypt_data_extended = {
    "YEAR": [2017, 2018, 2019, 2020, 2021, 2022, 2023],
    "FDI": [3.10, 2.88, 2.97, 1.60, 1.90, 2.44, 2.51],
    "UNEMPLOY": [11.5, 10.9, 9.594, 9.785, 12.432, 11.445, 10.454]
}

nigeria_data_extended = {
    "YEAR": [2017, 2018, 2019, 2020, 2021, 2022, 2023],
    "FDI": [8.0, 7.5, 7.6800, 9.2400, 9.7870, 8.5540, 8.5432],
    "UNEMPLOY": [7.9, 8.1, 8.53, 9.71, 9.79, 12.45, 12.98]
}

# Convert to DataFrames
egypt_df = pd.DataFrame(egypt_data_extended)
nigeria_df = pd.DataFrame(nigeria_data_extended)
```

```python
def likelihood(params, X, y):
    beta0, beta1, sigma = params
    y_pred = beta0 + beta1 * X
    # The negative log-likelihood for a normal distribution
    nll = -np.sum(-0.5 * np.log(2 * np.pi * sigma ** 2) - ((y - y_pred) ** 2) / (2 * sigma ** 2))
    return nll
```

```python
# Initial guesses for beta0, beta1, and sigma
initial_params = [0, 0, 1]

# Extracting data
X_egypt = egypt_df['FDI'].values
y_egypt = egypt_df['UNEMPLOY'].values

# Perform MLE
result_egypt = minimize(likelihood, initial_params, args=(X_egypt, y_egypt), method='L-BFGS-B', bounds=[(-np.inf, np.inf), (-np.inf, np.inf), (1e-6, np.inf)])
params_egypt = result_egypt.x
print(f"Egypt MLE Results: beta0 = {params_egypt[0]}, beta1 = {params_egypt[1]}, sigma = {params_egypt[2]}")
```

```python
# Extracting data
X_nigeria = nigeria_df['FDI'].values
y_nigeria = nigeria_df['UNEMPLOY'].values

# Perform MLE
result_nigeria = minimize(likelihood, initial_params, args=(X_nigeria, y_nigeria), method='L-BFGS-B', bounds=[(-np.inf, np.inf), (-np.inf, np.inf), (1e-6, np.inf)])
params_nigeria = result_nigeria.x
print(f"Nigeria MLE Results: beta0 = {params_nigeria[0]}, beta1 = {params_nigeria[1]}, sigma = {params_nigeria[2]}")
```

```python
# Calculate log-likelihoods
log_likelihood_egypt = -likelihood(params_egypt, X_egypt, y_egypt)
log_likelihood_nigeria = -likelihood(params_nigeria, X_nigeria, y_nigeria)

print(f"Egypt Log-Likelihood: {log_likelihood_egypt}")
print(f"Nigeria Log-Likelihood: {log_likelihood_nigeria}")
```

```python
import pandas as pd
import numpy as np
from scipy.optimize import minimize

# Extended data for Egypt and Nigeria (2017-2023)
egypt_data_extended = {
    "YEAR": [2017, 2018, 2019, 2020, 2021, 2022, 2023],
    "FDI": [3.10, 2.88, 2.97, 1.60, 1.90, 2.44, 2.51],
    "UNEMPLOY": [11.5, 10.9, 9.594, 9.785, 12.432, 11.445, 10.454]
}

nigeria_data_extended = {
    "YEAR": [2017, 2018, 2019, 2020, 2021, 2022, 2023],
    "FDI": [8.0, 7.5, 7.6800, 9.2400, 9.7870, 8.5540, 8.5432],
    "UNEMPLOY": [7.9, 8.1, 8.53, 9.71, 9.79, 12.45, 12.98]
}

# Convert to DataFrames
egypt_df = pd.DataFrame(egypt_data_extended)
nigeria_df = pd.DataFrame(nigeria_data_extended)

def likelihood(params, X, y):
    beta0, beta1, sigma = params
    y_pred = beta0 + beta1 * X
    nll = -np.sum(-0.5 * np.log(2 * np.pi * sigma ** 2) - ((y - y_pred) ** 2) / (2 * sigma ** 2))
    return nll

# Initial guesses for beta0, beta1, and sigma
initial_params = [0, 0, 1]

# Extracting data for Egypt
X_egypt = egypt_df['FDI'].values
y_egypt = egypt_df['UNEMPLOY'].values

# Perform MLE for Egypt
result_egypt = minimize(likelihood, initial_params, args=(X_egypt, y_egypt), method='L-BFGS-B', bounds=[(-np.inf, np.inf), (-np.inf, np.inf), (1e-6, np.inf)])
params_egypt = result_egypt.x
print(f"Egypt MLE Results: beta0 = {params_egypt[0]}, beta1 = {params_egypt[1]}, sigma = {params_egypt[2]}")

# Extracting data for Nigeria
X_nigeria = nigeria_df['FDI'].values
y_nigeria = nigeria_df['UNEMPLOY'].values

# Perform MLE for Nigeria
result_nigeria = minimize(likelihood, initial_params, args=(X_nigeria, y_nigeria), method='L-BFGS-B', bounds=[(-np.inf, np.inf), (-np.inf, np.inf), (1e-6, np.inf)])
params_nigeria = result_nigeria.x
print(f"Nigeria MLE Results: beta0 = {params_nigeria[0]}, beta1 = {params_nigeria[1]}, sigma = {params_nigeria[2]}")

# Calculate log-likelihoods
log_likelihood_egypt = -likelihood(params_egypt, X_egypt, y_egypt)
log_likelihood_nigeria = -likelihood(params_nigeria, X_nigeria, y_nigeria)

print(f"Egypt Log-Likelihood: {log_likelihood_egypt}")
print(f"Nigeria Log-Likelihood: {log_likelihood_nigeria}")
```

```python
import statsmodels.api as sm
from scipy.stats import chi2

# Define the regression models for Nigeria and Egypt
nigeria_model = sm.OLS(nigeria_df['UNEMPLOY'], sm.add_constant(nigeria_df[['FDI', 'RGDP', 'INTEREST', 'INF']])).fit()
egypt_model = sm.OLS(egypt_df['UNEMPLOY'], sm.add_constant(egypt_df[['FDI', 'RGDP', 'INTEREST', 'INF']])).fit()

# Calculate the likelihoods of each model
nigeria_likelihood = nigeria_model.llf
egypt_likelihood = egypt_model.llf

# Perform the likelihood ratio test
likelihood_ratio = -2 * (nigeria_likelihood - egypt_likelihood)
degrees_of_freedom = len(nigeria_model.params) - len(egypt_model.params)
p_value = 1 - chi2.cdf(likelihood_ratio, degrees_of_freedom)

print("Likelihood Ratio Test:")
print("Likelihood Ratio:", likelihood_ratio)
print("Degrees of Freedom:", degrees_of_freedom)
print("p-value:", p_value)
```
